# Differentials

## Introduction

In this notebook, we will discuss the concept of differentials. Differentials are a way to approximate the change in a function as its input changes. They are used in calculus to approximate the change in a function as its input changes by a small amount. The differential of a function is defined as the derivative of the function times the change in the input. In this notebook, we will discuss the concept of differentials and how they can be used to approximate the change in a function.

differentials character "dx" is used to represent the change in the input of a function. The differential of a function f(x) is defined as:

df(x) = f'(x)dx

where f'(x) is the derivative of the function f(x) and dx is the change in the input of the function. The differential df(x) represents the change in the output of the function as the input changes by a small amount dx.

#### key concept
- it is identic with small change in the function
- `dx . dx` (a bit of a bit) is negligible
- `dx/dy` represents the rate of change of y with respect to x (correlation between x and y)

![image.png](./photos/fig-5.png)

supposed we have `y = 180 cm` and `x = 19 cm` then if x grows by 1 cm, what is the change in y?

answer

1. We have to know the length of the triangle, which is `sqrt(180^2 + 19^2) = 181 cm`
2. The length of ladder will always be 181 cm. Then we can find the change in y by subtitude the x with `x + dx` where `dx = 1 cm`. It can be write like this. let say lader is `l`

$$
{l^2} = {x^2} + {y^2} \\
{l^2} = {(x+dx)^2} + {(y-dy)^2} \\
{(y-dy)^2} = {l^2} - {(x+dx)^2} \\
dy = -(\sqrt{l^2 - (x+dx)^2} - y)
$$

remember that `y = 180 cm`

$$
dy = -(\sqrt{181^2 - (19+1)^2} - 180) \\
dy = -(\sqrt{32761 - 400} - 180) \\
dy = -(\sqrt{32361} - 180) \\
dy = -(179.89 - 180) \\
dy = -(-0.11) cm \\
dy = 0.11 cm
$$

It means, if x grows by 1 cm, then y will decrease by 0.11 cm. Or it can be written as `dy/dx = 0.11/1`
