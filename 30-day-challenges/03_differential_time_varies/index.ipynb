# Differential Time Varies

### Key Concept

if y = distance and t = time, then the rate of change of y with respect to t is the speed of the object.

$$
v = y/t
$$

A velocity, scientifically expressed, is the rate at which a very small distance in any given direction is being passed over.

$$
v = \frac{dy}{dt}
$$

But, cause of the velocity `v` is not uniform, then it must be a change in velocity either increasing or else decreasing at any given time. This change in velocity is called acceleration.

$$
a = \frac{dv}{dt}
$$

We know that
$$
v = \frac{dy}{dt}
dv = d(\frac{dy}{dt})
$$

Then `a` supposed to be
$$
a = \frac{d^2y}{dt^2}
$$

# Introducing A Useful Dodge

$$
y = (x^2 + a^2)^\frac{3}{2}
$$

if we use from the concept that we have been learned before, we will convert it into u as some x function

$$
\frac{dy}{du} = \frac{3}{2}u^\frac{1}{2}
$$

Remember we still have this equation

$$
u = x^2 + a^2
\frac{du}{dx} = 2x
$$

Integrate it

$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} \\
\frac{dy}{dx} = \frac{3}{2}u^\frac{1}{2} \cdot 2x \\
\frac{dy}{dx} = 3x(x^2 + a^2)^\frac{1}{2}
$$


Now solve this

$$
y = \sqrt{\frac{1-x}{1+x}}
$$

remember concept of differential multiplication

$$
y = \frac{(1-x)^\frac{1}{2}}{(1+x)^\frac{1}{2}} \\
\frac{dy}{dx} = \frac{ (1+x)^\frac{1}{2} \cdot \frac{d(1-x)^\frac{1}{2}}{dx} - (1-x)^\frac{1}{2} \cdot \frac{d(1+x)^\frac{1}{2}}{dx}}{1+x} \\
$$

$$
\frac{d(1-x)^\frac{1}{2}}{dx} = -\frac{1}{2\sqrt{1-x}} \\
\frac{d(1+x)^\frac{1}{2}}{dx} = \frac{1}{2\sqrt{1+x}} \\
$$

hence

$$
\frac{dy}{dx} = \frac{ (1+x)^\frac{1}{2} \cdot -\frac{1}{2\sqrt{1-x}} - (1-x)^\frac{1}{2} \cdot \frac{1}{2\sqrt{1+x}}}{1+x} \\
\frac{dy}{dx} = \frac{ -\sqrt{1+x} + \sqrt{1-x}}{2(1+x)^\frac{3}{2}}
$$
