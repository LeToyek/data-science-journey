# Differential Multiplication

Suppose we have this differential equation:
$$
y = (x^2 + c) \cdot (ax^4+b)
$$

We wanna find the equation of (dy/dx) in terms of x and y. But the result will certainly **not** be $$ 2x \cdot 4ax^3 $$. We have to use the product rule of differentiation. In other word, do the algebraic multiplication first, then differentiate the result.

$$
y = (x^2 + c) \cdot (ax^4+b) \\
y = ax^6 + bx^2 + cax^4 + cb \\
\frac{dy}{dx} = 6ax^5 + 2bx + 4acx^3
$$

### Principal

Imagine that that equation is a multiplication of two `x` function. We can call it `u` and `v`. The product rule of differentiation is:
$$
y = u \cdot v \\
$$

Then we find the differential as before 
$$
y + dy = (u + du) \cdot (v + dv) \\
y + dy = u \cdot v + u \cdot dv + v \cdot du + du \cdot dv \\
$$

here $$ du \cdot dv $$ is negligible, due to smallness of the differential. So we have:

$$
y + dy = u \cdot v + u \cdot dv + v \cdot du \\
$$

remember that $$ u \cdot v = y $$ so we have:

$$
y + dy - y = u \cdot dv + v \cdot du \\
dy = u \cdot dv + v \cdot du \\
\frac{dy}{dx} = u \cdot \frac{dv}{dx} + v \cdot \frac{du}{dx} \\
$$

Let's apply this principle to the equation above. We have:
$$
u = x^2 + c \\
v = ax^4 + b \\
$$

Then we find the differential as before
$$
\frac{dy}{dx} = (x^2 + c) \cdot 4ax^3 + (ax^4 + b) \cdot 2x \\
\frac{dy}{dx} = 4ax^5 + 4acx^3 + 2ax^5 + 2bx \\
\frac{dy}{dx} = 6ax^5 + 2bx + 4acx^3 \\
$$

**Division**

suppose we have

$$
y = \frac{u}{v}
$$
  
In somecases, the `u` and `v` hhave no any common factor so it is not possible to simplify the equation. In this case, we can use the quotient rule of differentiation. The quotient rule is:

$$
y = \frac{u}{v} \\
y + dy = \frac{u + du}{v + dv} \\
y + dy \cdot v + dv = u + du \\
vy + ydv + vdy + dy \cdot dv = u + du \\
$$

always remember that $$ dy \cdot dv $$ is negligible, due to smallness of the differential. So we have:

$$
vy + ydv + vdy = u + du \\
\frac{u}{v}v + \frac{u}{v}dv + vdy = u + du \\
u + \frac{udv}{v} + vdy = u + du \\
\frac{udv}{v} + vdy = du \\
udv + v^2dy = vdu \\
v^2dy = vdu - udv \\
dy = \frac{vdu - udv}{v^2} \\
$$

and Finally:

$$
\frac{dy}{dx} = \frac{v\frac{du}{dx} - u\frac{dv}{dx}}{v^2} \\
$$

### Example

Suppose we have this differential equation:
$$
y = \frac{bx^5+c}{x^2 + a}
$$

Just applpy the quotient rule of differentiation:
$$
u = bx^5 + c \\
v = x^2 + a \\
$$

Then we find the differential as before
$$
\frac{dy}{dx} = \frac{(x^2 + a) \cdot 5bx^4 - (bx^5 + c) \cdot 2x}{(x^2 + a)^2} \\
\frac{dy}{dx} = \frac{5abx^6 + 5bx^4a - 2bx^6 - 2cx}{(x^2 + a)^2} \\
\frac{dy}{dx} = \frac{3abx^6 + 5bx^4a - 2cx}{(x^2 + a)^2} \\
$$


## Hot quiz

A reservoir of square cross-section has sides sloping at an angle of 45 degree with the vertical. The side of the bottom is 200 feet. Find an expression for the quantity pouring in our out when the depth of water varies by foot.

### Solution

![image.png](./photos/reservoir.png)

That is the figure of the reservoir. We know that if the height is `h` feet than the width of top reservoir is `200 + tan(45) . h . 2`. 

We know that
$$
tan(45) = 1
$$

so the width of top reservoir is `200 + h . 2`.
assume a = bottom area and A = top area

$$
a = 200^2 \\
b = (200 + 2h)^2 \\
$$

We use the frustum of pyramid formula to find the volume of the reservoir:

$$
V = \frac{1}{3} \cdot h \cdot (a + A + \sqrt{aA}) \\
V = \frac{1}{3} \cdot h \cdot (200^2 + (200 + 2h)^2 + \sqrt{200^2 \cdot (200 + 2h)^2}) \\
V = \frac{1}{3} \cdot h \cdot (200^2 + (200 + 2h)^2 + 200 \cdot (200 + 2h)) \\
V = \frac{1}{3} \cdot h \cdot (200^2 + 200^2 + 400h + 400h + 4h^2 + 200 \cdot 200 + 200 \cdot 2h) \\
V = \frac{1}{3} \cdot h \cdot (40000 + 40000 + 800h + 4h^2 + 40000 + 400h) \\
V = \frac{1}{3} \cdot h \cdot (120000 + 1200h + 4h^2) \\
V = \frac{1}{3} \cdot (120000h + 1200h^2 + 4h^3) \\
V = 40000h + 400h^2 + \frac{4}{3}h^3 \\
$$

and finally

$$
\frac{dV}{dh} = 40000 + 800h + 4h^2 \\
$$

suppose h = 12

$$
\frac{dV}{dh} = 40000 + 800 \cdot 12 + 4 \cdot 12^2 \\
\frac{dV}{dh} = 40000 + 9600 + 576 \\
\frac{dV}{dh} = 50576 \\
$$

$$

find change of E per degree at 15,20,25 degree

$$
E = 1.434 - 1.434 \cdot 0.000814(t-15) + 1.434 \cdot 0.0000007(t-15)^2 \\
E = 1.434 - 0.001167276 \cdot (t-15) + 0.0000007 \cdot (t^2 - 30t + 225) \\

$$

and finally

$$
\frac{dE}{dt} = -0.001167276 + 0.0000007 \cdot 2t - 0.000021 \\
\frac{dE}{dt} = -0.001188276 + 0.0000014t \\
